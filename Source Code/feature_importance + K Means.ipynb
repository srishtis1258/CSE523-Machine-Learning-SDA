```python
KNNdata = pd.read_csv("Dataset/All_Recovery_files.csv")
KNNdata = KNNdata.drop(["Sleep Onset","Sleep Onset","Wake Onset","Last Nap End Time","timezone_offset","Nap Time in Bed","Nap Disturbances","Nap REM","Nap SWS","Nap Light","Nap Wake","Nap Missing Data"],axis=1)
KNNdata.drop(KNNdata.iloc[:, 0:4], inplace = True, axis = 1) 

KNNdata = KNNdata.iloc[:,0:len(KNNdata.columns)-2]

KNNdata
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=3)
```

```python
df_filled = imputer.fit_transform(KNNdata)
New_KNNdata = pd.DataFrame(df_filled, columns = KNNdata.columns)
New_KNNdata.to_csv("KNN_entireDF.csv")
New_KNNdata
```
